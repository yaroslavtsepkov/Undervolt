```
import pandas as pd
import matplotlib.pyplot as plt
```

```
profile = pd.DataFrame(data=pd.read_csv('/content/sample_data/profile.csv'))
stock = pd.DataFrame(data=pd.read_csv('/content/sample_data/stock.csv'))
```

```
data_stock = pd.DataFrame(data=stock)
data_profile = pd.DataFrame(data=profile)
```

```
data_stock.head(5)
```

```
data_profile.head(5)
```

```
print('FPS AVG stock:',data_stock['FPS'].mean())
print('Temp AVG stock:',data_stock['celcius'].mean())
print('FPS AVG profile:',data_profile['FPS'].mean())
print('Temp AVG profile:',data_profile['celcius'].mean())
print('Decrease in temperature AVG', data_stock['celcius'].mean() - data_profile['celcius'].mean())
```

```
fig, ax = plt.subplots(2, 1, figsize=(15,10))
ax[0].plot(data_stock['seconds'], data_stock['celcius'],linewidth = 2, label='stock')
ax[0].plot(data_profile['seconds'], data_profile['celcius'],linewidth = 2,linestyle = 'solid',  label='profile')
ax[0].set_xlabel('time', fontsize=16)
ax[0].set_ylabel('celcius', fontsize=16)
ax[0].legend(loc='lower right', fontsize=14)
ax[0].grid(True)

ax[1].plot(data_stock['seconds'], data_stock['FPS'],linewidth = 2, label = 'stock')
ax[1].plot(data_profile['seconds'], data_profile['FPS'],linewidth = 2, label = 'profile')
ax[1].set_xlabel('time', fontsize=16)
ax[1].set_ylabel('FPS', fontsize=16)
ax[1].legend(loc='lower right', fontsize=14)
ax[1].grid(True)

fig.savefig(fname='graph.svg')
fig.show()
```
