```
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```
profile = pd.DataFrame(data=pd.read_csv('/content/profile.csv'))
stock = pd.DataFrame(data=pd.read_csv('/content/stock.csv'))
```

```
data_stock = pd.DataFrame(data=stock)
data_profile = pd.DataFrame(data=profile)
```

```
data_stock
```

```
data_profile
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=data_stock['seconds'], y=data_stock['celcius'],
                    mode='lines',
                    name='stock'))
fig.add_trace(go.Scatter(x=data_profile['seconds'], y=data_profile['celcius'],
                    mode='lines',
                    name='profile'))
fig.update_layout(title='Temp GTX 1060 3G Palit Dual',
                   xaxis_title='Seconds',
                   yaxis_title='Temperature (degrees C)')

fig.show()
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=data_stock['seconds'], y=data_stock['FPS'],
                    mode='lines',
                    name='stock'))
fig.add_trace(go.Scatter(x=data_profile['seconds'], y=data_profile['FPS'],
                    mode='lines',
                    name='profile'))
fig.update_layout(title='Temp GTX 1060 3G Palit Dual',
                   xaxis_title='Seconds',
                   yaxis_title='FPS')

fig.show()
```
